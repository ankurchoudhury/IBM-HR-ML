```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = 'white')
sns.set(style = 'whitegrid', color_codes = True)
```

```python
df = pd.read_csv(r'dataset.csv')
df.drop(['EmployeeCount', 'Over18', 'StandardHours', 'EmployeeNumber'],
        axis = 1,
        inplace = True)
df.head()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.countplot(data = df, x = 'Attrition')
```

The figure indicates that the dataset is highly imbalanced. The number of attrition cases is very small.<br>
Though this is a true reflection of the real world, a balanced dataset is important to create a proper ML model.
