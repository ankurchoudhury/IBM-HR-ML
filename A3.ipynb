```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'dataset.csv')
df.drop(['EmployeeCount', 'Over18', 'StandardHours', 'EmployeeNumber'],
        axis = 1,
        inplace = True)
df.head()
```

```python
X = pd.DataFrame()
X['Attrition'] = np.where(df['Attrition'] == 'Yes', 1, 0)
X['Female'] = np.where(df['Gender'] == 'Female', 1, 0)
X['OverTime'] = np.where(df['OverTime'] == 'Yes', 1, 0)
```

```python
dummy = pd.get_dummies(df['BusinessTravel'])
dummy.drop(['Non-Travel'], axis = 1, inplace = True)
dummy.columns = ['a.Travel_Frequently',
                 'a.Travel_Rarely']
X = pd.concat([X, dummy], axis = 1)
```

```python
dummy = pd.get_dummies(df['Department'])
dummy.drop(['Research & Development'], axis = 1, inplace = True)
dummy.columns = ['b.HR',
                 'b.Sales']
X = pd.concat([X, dummy], axis = 1)
```

```python
dummy = pd.get_dummies(df['MaritalStatus'])
dummy.drop(['Divorced'], axis = 1, inplace = True)
dummy.columns = ['c.Married',
                 'c.Single']
X = pd.concat([X, dummy], axis = 1)
```

```python
dummy = pd.get_dummies(df['EducationField'])
dummy.drop(['Other'], axis = 1, inplace = True)
dummy.columns = ['d.Human_Resources',
                 'd.Life_Sciences',
                 'd.Marketing',
                 'd.Medical',
                 'd.Technical_Degree']
X = pd.concat([X, dummy], axis = 1)
```

```python
dummy = pd.get_dummies(df['JobRole'])
dummy.drop(['Research Director'], axis = 1, inplace = True)
dummy.columns = ['e.Healthcare_Rep',
                 'e.Human_Resources',
                 'e.Lab_Technician',
                 'e.Manager',
                 'e.Mfg_Director',
                 'e.Research_Scientist',
                 'e.Sales_Executive',
                 'e.Sales_Rep']
X = pd.concat([X, dummy], axis = 1)
```

```python
df_num = df.select_dtypes(['number'])
X = pd.concat([X, df_num], axis = 1)
X.head()
```

```python
X.to_csv(r'dataset_2.csv', index = False)
```
